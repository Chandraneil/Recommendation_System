```python
import pandas as pd
import numpy as np
from transformers4rec.preprocessing import OneHotEncoder, SequencePreprocessor

# Load data
df = AllData

# Convert user and item IDs to sequential integers
user_encoder = OneHotEncoder()
user_ids = user_encoder.fit_transform(df['user'])

item_encoder = OneHotEncoder()
item_ids = item_encoder.fit_transform(df['item'])

# Convert ratings to numpy array
ratings = df['rating'].values.astype(np.float32)

# Create sequence data
sequence_data = np.stack([user_ids, item_ids], axis=1)

# Define sequence preprocessor
preprocessor = SequencePreprocessor()

# Preprocess sequence data
preprocessed_data = preprocessor.fit_transform(sequence_data, ratings)

# Save preprocessed data to disk
preprocessor.save('preprocessor.pkl')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df

# Plot 1: Histogram of a numerical variable
plt.figure(figsize=(8, 6))
sns.histplot(df['liked'], kde=True)
plt.title('Histogram of numerical variable')

# Plot 2: Scatter plot of two numerical variables
plt.figure(figsize=(8, 6))
sns.scatterplot(x='numerical_variable_1', y='numerical_variable_2', data=df)
plt.title('Scatter plot of two numerical variables')

# Plot 3: Box plot of a categorical and numerical variable
plt.figure(figsize=(8, 6))
sns.boxplot(x='categorical_variable', y='numerical_variable', data=df)
plt.title('Box plot of a categorical and numerical variable')

# Plot 4: Bar chart of a categorical variable
plt.figure(figsize=(8, 6))
sns.countplot(x='categorical_variable', data=df)
plt.title('Bar chart of a categorical variable')

# Plot 5: Line plot of a numerical variable over time
plt.figure(figsize=(8, 6))
sns.lineplot(x='time_variable', y='numerical_variable', data=df)
plt.title('Line plot of a numerical variable over time')

# Plot 6: Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Heatmap of the correlation matrix')

plt.show()
```
